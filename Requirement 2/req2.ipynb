```python
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float

Base = declarative_base()

class Resale(Base):
    __tablename__ = 'resales'
    transaction_id = Column(Integer, primary_key=True)
    month = Column(Integer)
    year = Column(Integer)
    town = Column(String)
    flat_type = Column(String)
    block = Column(String)
    street_name = Column(String)
    storey_range_min = Column(Integer)
    storey_range_max = Column(Integer)
    floor_area_sqm = Column(Float)
    flat_model = Column(String)
    lease_commence_date = Column(Integer)
    resale_price = Column(Float)
    end_of_lease_year = Column(Integer)
    years_remaining = Column(Integer)
    
    def __repr__(self):
        return "%s %s %s %s %s %s %s %s %s %s %s %s %s %s %s" %(self.transaction_id, self.month, self.year, self.town, self.flat_type,
                        self.block, self.street_name, self.storey_range_min, self.storey_range_max, self.floor_area_sqm,
                        self.flat_model, self.lease_commence_date, self.resale_price, self.end_of_lease_year,
                        self.years_remaining)
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///resaleflats.db', echo=True)
Base.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import sessionmaker

engine = create_engine('sqlite:///resaleflats.db', echo=True)
Session = sessionmaker(bind=engine)
session = Session()
```

```python
result = session.query(Resale)
for r in result:
    print(r)
```

```python
def orm_add(session, row):
    transaction_id = row[0]
    month = row[1]
    year = row[2]
    town = row[3]
    flat_type = row[4]
    block = row[5]
    street_name = row[6]
    storey_range_min = int(row[7].split('TO')[0])
    storey_range_max = int(row[7].split('TO')[1])
    floor_area_sqm = row[8]
    flat_model = row[9]
    lease_commence_date = row[10]
    resale_price = row[11]
    end_of_lease_year = row[12]
    years_remaining = row[13]
    
    session.add(Resale(transaction_id=transaction_id, month=month, year=year, town=town, flat_type=flat_type, block=block,
               street_name=street_name, storey_range_min=storey_range_min, storey_range_max=storey_range_max, floor_area_sqm=floor_area_sqm, flat_model=flat_model,
               lease_commence_date=lease_commence_date, resale_price=resale_price, end_of_lease_year=end_of_lease_year, years_remaining=years_remaining))
```

```python
import csv

line_count = 0

with open('transformed_resale_flat.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    header = next(csv_reader)
    
    for row in csv_reader:
        orm_add(session, row)
        line_count += 1
        
    session.commit()
```

```python
# Question 3b(i)

def top3_towns_per_month(df):
    new_df = df.groupby(['year','month','town']).size().reset_index(name='count')
    level1_groupby_df = new_df.groupby(['year','month','town']).agg({'count':sum})
    level2_groupby_df = level1_groupby_df['count'].groupby(['year','month'], group_keys=False)
    top3_df = level2_groupby_df.apply(lambda x: x.sort_values(ascending=False).head(3))

    return top3_df

top3_towns_per_month(df)
```
